```python
# -------- 4. Compute counts per company --------
rows = []
for _, row in df.iterrows():
    company = str(row.get("company_name"))
    firmo = count_signals(row, FIRMO_COLS)
    web   = count_signals(row, WEB_COLS)
    tech  = count_signals(row, TECH_COLS)
    fin   = count_signals(row, FIN_COLS)
    growth = count_signals(row, GROWTH_COLS)
    total  = firmo + web + tech + fin + growth

    rows.append({
        "company_name": company,
        "firmo_signals": firmo,
        "web_signals": web,
        "tech_signals": tech,
        "fin_signals": fin,
        "growth_signals": growth,
        "total_signals": total,
    })

signals_df = pd.DataFrame(rows)
signals_df
```

```python
print("=== Signal counts per company ===")
print(signals_df.to_string(index=False))

# -------- 5. Dict you can paste into Streamlit --------
STATIC_SIGNAL_COUNTS = (
    signals_df
    .set_index("company_name")[["firmo_signals","web_signals","tech_signals","fin_signals","growth_signals"]]
    .to_dict(orient="index")
)

print("\nSTATIC_SIGNAL_COUNTS =")
print(STATIC_SIGNAL_COUNTS)
```
